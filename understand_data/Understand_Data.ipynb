```python
import pandas as pd
iris = pd.read_csv('iris.csv')
```

```python
print(iris.head(5))
```

```python
iris.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))

plt.xlabel("Species", fontsize=16)
plt.ylabel("Sepal Length", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
sns.boxplot(x='Species', y='SepalLengthCm', data=iris)
plt.show()
```

```python

sns.boxplot(x='Species', y='SepalLengthCm', data=iris)
```

```python
sns.barplot(x='Species', y='SepalLengthCm', data=iris)
```

```python
iris['SepalLengthCm'].plot.hist( bins=10)
```

```python
iris['SepalLengthCm'].plot.density()
```

```python
sns.displot(iris, x="SepalLengthCm", hue="Species", bins=5)
```

```python
from plotnine import *
```
