## Box Plot using seaborn library

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))

plt.xlabel("Species", fontsize=16)
plt.ylabel("Sepal Length", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
sns.boxplot(x='Species', y='SepalLengthCm', data=iris)
# plt.show()
```

## Histogram Plots with bin

```python
sns.displot(iris, x="SepalLengthCm", hue="Species", bins=5)
```

## Density Plot

```python
sns.displot(iris, x="SepalLengthCm", hue="Species",  kind="kde")
```

## Petal Length Density Plot

```python
sns.displot(iris, x="PetalLengthCm", hue="Species",  kind="kde")
```
